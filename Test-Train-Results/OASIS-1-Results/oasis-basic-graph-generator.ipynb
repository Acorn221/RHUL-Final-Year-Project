```python
# Import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import json
import pandas as pd
import os


# define the path to the directory containing the JSON files
path = r'C:\Active-Projects\RHUL-FYP\PROJECT\Test-Train-Results\OASIS-1-Results\models'

# Import the JSON files
json_files = [pos_json for pos_json in os.listdir(path) if pos_json.endswith('.json')]

for file in json_files:
		with open(os.path.join(path, file)) as f:

				title = file.split('.json')[0] + ' - Transfer Learning With The OASIS-1 Dataset'
				# Create a pandas dataframe from the JSON file
				df = pd.read_json(f)
				history = df['history']
				# Plot the data
				plt.plot(history['loss'], label='loss')
				plt.plot(history['val_loss'], label='val_loss')
				plt.xlabel('Epoch')
				plt.ylabel('Loss')
				plt.title(title)
				plt.legend()
				plt.grid(True)
				plt.show()
				plt.plot(history['categorical_accuracy'], label='categorical_accuracy')
				plt.plot(history['val_categorical_accuracy'], label='val_categorical_accuracy')
				plt.xlabel('Epoch')
				plt.ylabel('Accuracy')
				plt.title(title)
				plt.legend()
				plt.grid(True)
				plt.show()
				

```

```python
from matplotlib.pyplot import figure

plt.rcParams["figure.figsize"] = (15,5)

for file in json_files:
		with open(os.path.join(path, file)) as f:
				# Create a pandas dataframe from the JSON file
				name = file.split('.json')[0]
				df = pd.read_json(f)
				history = df['history']
				# Plot the data
				plt.plot(history['val_loss'], label=name)
				#plt.plot(history['val_loss'], label='val_loss_'+name)
				
plt.ylim(0, 1.5)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title("Transfer Learning With The OASIS-1 Dataset Validation Loss")
plt.legend()
plt.grid(True)
ax = plt.subplot(111)
#Move the legend to under the graph
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          ncol=5, fancybox=True, shadow=True)

plt.show()
```
